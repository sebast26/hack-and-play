```kotlin
import io.ktor.client.request.parameter

val forecast = http.get("https://api.open-meteo.com/v1/forecast") {
    parameter("latitude", minLat + (maxLat - minLat) / 2)
    parameter("longitude", minLon + (maxLon - minLon) / 2)
    parameter("hourly", "wind_speed_10m,wind_direction_10m")
    parameter("wind_speed_unit", "kn")
}.deserializeJson()

forecast
```

```kotlin
fun forecastAtPoint(latitude: Double, longitude: Double): Forecast =
    http.get("https://api.open-meteo.com/v1/forecast") {
        parameter("latitude", latitude)
        parameter("longitude", longitude)
        parameter("hourly", "wind_speed_10m,wind_direction_10m")
        parameter("wind_speed_unit", "kn")
    }.body()
```

```kotlin
infix fun ClosedFloatingPointRange<Double>.step(step: Double) =
    generateSequence(start) { prev ->
        val next = prev + step
        if (next <= endInclusive) next else null
    }.toList()

val wind = (minLat..maxLat step 0.1)
    .flatMap { lat -> (minLon..maxLon step 0.1).map { lat to it } }
    .map { (lat, lon) ->
        forecastAtPoint(lat, lon)
    }
```

```kotlin
val windDF = wind.toDataFrame(maxDepth = 2)
```

```kotlin
val windForecast = windDF
    .explode { hourly.colsAtAnyDepth() }
    .select { longitude and latitude and hourly }
    .flatten { hourly }
windForecast
```

```kotlin
data class State(
    val currentPoint: String,
    val currentTime: LocalDateTime,
    val totalDistance: Double,
    val usedRoutes: List<Int>,
)
```

```kotlin
val windInPoints =
    boeien18052025.df.map { name to forecastAtPoint(geometry.coordinate.y, geometry.coordinate.x).hourly }.toMap()
```

```kotlin
val timeOfStart = LocalDateTime(2025, 5, 23, 18, 0)
val finishTime = LocalDateTime(2025, 5, 24, 18, 0)
val finishNode = "WV19"
val states = rakken23052025.filter { maxAantal == 1 }.start.toSet().map { State(it, timeOfStart, 0.0, emptyList()) }
    .toMutableList()
val finishStates = mutableListOf<State>()
```

```kotlin
import kotlinx.datetime.TimeZone
import kotlinx.datetime.toInstant
import kotlinx.datetime.toLocalDateTime
import kotlin.time.Duration

operator fun LocalDateTime.plus(duration: Duration): LocalDateTime {
    val zone = TimeZone.currentSystemDefault()
    val instant = this.toInstant(zone)
    val newInstant = instant + duration
    return newInstant.toLocalDateTime(zone)
}

fun calculateTrueWindAngle(
    currentState: State,
    currentWindDir: Int,
    start: String,
    startPoint: Point,
    endPoint: Point
): Double {

    val dx = endPoint.x - startPoint.x
    val dy = endPoint.y - startPoint.y

    var angleRad = atan2(dx, dy)
    var heading = Math.toDegrees(angleRad)
    if (currentState.currentPoint != start) heading += 180
    if (heading < 0) heading += 360

    return (abs(heading - currentWindDir) % 360).let {
        if (it > 180) 360 - it else it
    }
}
```

```kotlin
fun boatSpeed(twa: Double, windSpeed: Double): Double {
    return gathered.polars[windSpeed.toInt()].boatSpeed[twa.toInt()].toDouble()
}
```

```kotlin
while (states.isNotEmpty()) {
    val currentState = states.removeFirst()

    // getting wind
    val i = windInPoints[currentState.currentPoint]!!.time.indexOfFirst { it > currentState.currentTime.toString() }
    val currentWindSpeed = windInPoints[currentState.currentPoint]!!.windSpeed10m[i]
    val currentWindDir = windInPoints[currentState.currentPoint]!!.windDirection10m[i]

    //getting next points
    val nexts = segments.df
        .filter { start == currentState.currentPoint || eind == currentState.currentPoint }
        .filter { currentState.usedRoutes.count { it == id } < maxAantal }
        .add("TWA") {
            calculateTrueWindAngle(
                currentState,
                currentWindDir,
                start,
                geometry.startPoint,
                geometry.endPoint
            )
        }
        .filter { "TWA"<Double>() in 80.0..165.0 }

    // adding new states
    val nextStates = nexts.map {
        val newTime = currentState.currentTime + (it.afstand / boatSpeed("TWA"(), currentWindSpeed)).hours
        State(
            if (currentState.currentPoint == it.start) it.eind else it.start,
            newTime,
            currentState.totalDistance + it.afstand,
            currentState.usedRoutes + it.id,
        )
    }.forEach {
        if (it.currentTime < finishTime) {
            states.add(it)
            if (it.currentTime + 1.hours > finishTime && it.currentPoint == finishNode) {
                finishStates.add(it)
            }
        }
    }
}
```

```kotlin
val resultSegments = finishStates
    .filter { it.currentTime + 30.minutes < finishTime }
    .sortedByDescending { it.totalDistance }
    .first()
    .usedRoutes.toDataFrame("segment")
    .addId()
    .join(segments.df) { "segment" match right.id }
    .toGeo()

resultSegments
```
